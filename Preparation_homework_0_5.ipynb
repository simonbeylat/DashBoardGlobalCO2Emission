```python
Total = go.Scatter(
    x = df_Golbal['Year'],
    y = df_Golbal['Total'],
    name= 'Total'
)
Coal = go.Scatter(
    x = df_Golbal['Year'],
    y = df_Golbal['Coal'],
    name= 'Coal'
)
Oil = go.Scatter(
    x = df_Golbal['Year'],
    y = df_Golbal['Oil'],
    name= 'Oil'
)
Gas = go.Scatter(
    x = df_Golbal['Year'],
    y = df_Golbal['Gas'],
    name= 'Gas'
)
Cement = go.Scatter(
    x = df_Golbal['Year'],
    y = df_Golbal['Cement'],
    name= 'Cement'
)
Flaring = go.Scatter(
    x = df_Golbal['Year'],
    y = df_Golbal['Flaring'],
    name= 'Flaring'
)
Other = go.Scatter(
    x = df_Golbal['Year'],
    y = df_Golbal['Other'],
    name= 'Other'
)
data = go.Data([Total, Coal, Oil, Gas, Cement,Flaring,Other])

layout = go.Layout(
    title = "Global CO2 Emission",
    xaxis = go.XAxis(
        title = "Year",
    ),
    yaxis = go.YAxis(
        title = "CO2 Emission",
    )
)

figure = go.Figure(data=data, layout=layout)
plotly.offline.iplot(figure)
```

## 2D plot with post-processing 

```python
df.loc[df.Country == 'USA', 'Country'] = 'United States'
```

```python
df_continent = pd.read_csv('data/Countries-Continents.csv',encoding='latin-1')
```

```python
df2 = pd.merge(
  df, 
  df_continent, 
  left_on='Country', 
  right_on='Country',
  how="left",
)
```

```python
df_merged =df2.loc[ df2.Continent.isin(['Asia', 'Europe', 'Africa','North America', 'South America','Oceania'])]
```

```python
fig = px.bar(df_merged, x='Continent', y='Total',animation_frame='Year',hover_name='Country')
fig.show()
```

```python
fig = px.line(df_merged,x=df_merged['Year'],y=df_merged['Total'],line_group='Continent',color='Continent')
fig.show()
```

```python
df_without_Total = df.loc[df.Country != 'Global']
```
